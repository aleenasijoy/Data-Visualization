```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset('penguins')
print(data)
```

```python
data.isnull().sum()
```

```python
a=data['bill_length_mm'].mode().values[0]
```

```python
data['bill_length_mm']=data['bill_length_mm'].replace(np.NaN,a)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
v=LabelEncoder()
x=['species','island','sex']
for i in x:
    data[i]=v.fit_transform(data[i])
print(data)
```

```python
from sklearn.preprocessing import StandardScaler
c=StandardScaler()
data[['body_mass_g']]=c.fit_transform(data[['body_mass_g']])
print(data)
```

```python
data_corr=data.corr()
```

```python
sns.heatmap(data_corr,annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Sample data with two features
data = [[1, 2], [3, 4], [5, 6]]

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit scaler to the data and transform the data
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```
